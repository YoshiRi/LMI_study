# LQG and LQR control

LQRは最適制御，ただLQGはノイズを考えられる。

Given LTI system 
$$
\dot{x} = A x + B_u u + B_\omega \omega \\
y = C_yx + D_y \omega \\
z = C_z x + D_z u
$$

and, the observer based controller

$$
\dot{\hat{x}} = A \hat{x} + B_u u + F(\hat{y}-y)\\
\hat{y}=C_y\hat{x}\\
u=K\hat{x}
$$

- $(A,B_u)$ stabilizable
- $(A,C_y)$ detectable
- $\omega(t)$ is Gaussian with covariance $W$
- $C_z^\top D_z = 0$, $B_\omega W D_y^\top=0$


Evaluate function is:
$$
J := \lim_{t \rightarrow \infty} E[z(t)^\top z(t)]
$$
